```python
# 3. Explain the following terms:
#a) The difference in np.array, np.asarray and np.asanyarray
#b) The difference between Deep copy and shallow copy


"""
Use np.array when you want to ensure you have a NumPy array, even if the input is already an array.
Use np.asarray when you want to avoid unnecessary copying of data and are okay with getting a view of the input data.
Use np.asanyarray when you want to preserve the type of the input, especially if it's a subclass of np.ndarray.

"""

import numpy as np

# Original data (Python list)
data = [1, 2, 3, 4, 5]

# np.array always creates a new array
arr1 = np.array(data)
print(arr1 is data)  # False (new array created)

# np.asarray creates a view if input is already a NumPy array
arr2 = np.asarray(arr1)
print(arr2 is arr1)  # True (view of the same data)

# np.asanyarray preserves input type if it's a subclass of np.ndarray
sub_class_array = np.ma.masked_array([1, 2, 3, 4, 5])
arr3 = np.asanyarray(sub_class_array)
print(isinstance(arr3, np.ma.masked_array))  # True (preserves subclass type)

# b)

import numpy as np

# Original array
original_array = np.array([[1, 2], [3, 4]])

print("Original array:")
print(original_array)

# Shallow copy using view
shallow_copy = original_array.view()

# Deep copy using copy
deep_copy = original_array.copy()

# Modify the original array
original_array[0, 0] = 99

# Changes in the shallow copy affect the original array
print("\nShallow copy:")
print(shallow_copy)

# Changes in the deep copy do not affect the original array
print("\nDeep copy:")
print(deep_copy)
```

```python
# 6. Create a 3D NumPy array of shape (3, 3, 3) containing random integers Petween 1 and 10. Perform the following operations:

# a) Find the indices of the maximum values along each depth level (third axis).

# b) Perform element-wise multiplication of between both array

import numpy as np

array_3d = np.random.randint(1, 11, size=(3, 3, 3))

print("Original 3D array:")
print(array_3d)

# Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array_3d, axis=2)

print("\nIndices of maximum values along each depth level:")
print(max_indices)

# Perform element-wise multiplication between both arrays
elementwise_multiplication = array_3d * max_indices[:, :, np.newaxis]

print("\nElement-wise multiplication between both arrays:")
print(elementwise_multiplication)

```

```python
# 7. Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column.

import pandas as pd 
import numpy as np
df = pd.read_csv('People Data.csv')

def clean_phone(phone):
    if pd.isna(phone) or phone == '':
        return np.nan
    # Remove non-numeric characters
    cleaned_phone = ''.join(filter(str.isdigit, phone))
    # Return NaN if cleaned_phone is empty, otherwise return the cleaned phone number
    return np.nan if cleaned_phone == '' else int(cleaned_phone)

# Apply the cleaning function to the 'Phone' column
df['Phone'] = df['Phone'].apply(clean_phone)

# Convert the 'Phone' column to Int64 to handle integers with NaNs
df['Phone'] = df['Phone'].astype('Int64')

# Configure pandas to display numbers in full without scientific notation
pd.set_option('display.float_format', '{:.0f}'.format)

# Display the DataFrame and its data types
print("DATAFRAME : \n",df)
print("\nData Types :\n",df.dtypes)



```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
#Wave Function'

from bokeh.plotting import figure, show

x = list(range(100))

y = [np.sin(i) for i in x]

p = figure(title="Sine Wave Function", y_range=(-1.1, 1.1))

p.line(x, y, line_width=2)

p.grid.grid_line_color = None

show(p)
```

```python
#19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'

from bokeh.plotting import figure, show
from bokeh.palettes import Spectral8

data = {
    "categories": list("ABCDEFGH"),
    "values": np.random.randint(low=1, high=100, size=8),
    "colors": Spectral8
}

p = figure(title="Bar Chart", y_range=(0, data["values"].max() + 10))

p.vbar(x="categories", top="values", width=0.5, color="colors", source=data)

show(p)
```

```python
#20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

import plotly.graph_objects as go

x = list(range(100))
y = [np.random.randint(1, 100) for _ in range(100)]

fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines'))

fig.update_layout(
    title="Simple Line Plot",
    xaxis_title="X Axis",
    yaxis_title="Y Axis",
)

fig.show()
```

Q5. Show a pie chart to represent the percentage distribution of different sections in the array `sections`.
sections = ['Section A', 'Section B', 'Section C', 'Section D']

```python
sections = ['Section A', 'Section B', 'Section C', 'Section D']
sizes = [25, 30, 15, 30]
explode=(0.1,0.1,0.1,0.0)
plt.pie(sizes,labels=sections,autopct = '%1.1f%%', explode=explode, shadow=True)
plt.title("Pie Chart")
```

Q3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values

```python
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Value': np.random.randint(1, 100, size=5)  # Random numerical values for each category
}
df = pd.DataFrame(data)

# Plotting with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Value', data=df, palette='viridis')
plt.title('Comparison of Categories Based on Numerical Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.
```

```python
# Generate a dataset with categories and numerical values
np.random.seed(0)  # for reproducibility
categories = ['A', 'B', 'C', 'D', 'E']
num_values = 100
data = {
    'Category': np.random.choice(categories, num_values),
    'Value': np.random.normal(loc=50, scale=10, size=num_values)  # normally distributed values
}
df = pd.DataFrame(data)

# Visualize the distribution of the numerical variable across different categories
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Distribution of Numerical Variable Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```
